```python
def quantify_salary(sal_string):
    if sal_string == 'low':
        return 1
    elif sal_string == 'medium':
        return 2
    else:
        return 3
```

```python
def print_heatmap(column_value, column, dataframe):
    print(column_value, '\n')
    slicer = dataframe[dataframe[column] == column_value]
    corr = slicer.corr()
    print(corr)
    sns.heatmap(corr, 
                xticklabels=corr.columns.values,
                yticklabels=corr.columns.values)
    sns.plt.title('Heatmap of {} Correlation Matrix'.format(str(column_value).upper()))
    plt.show()
    print('\n')
    plt.clf()
```

```python
new_df.shape
```

```python
new_df.columns
```

```python
for d in departments:
    print_heatmap(d, 'department', new_df)
```

```python
print_heatmap(1, 'sal__low', new_df)
```

```python
for department in departments:
    print(department, '\n')
    slicer = new_df[new_df['department'] == department]
    corr = slicer.corr()
    sns.heatmap(corr, 
                xticklabels=corr.columns.values,
                yticklabels=corr.columns.values)
    sns.plt.title('Heatmap of {} Correlation Matrix'.format(department.upper()))
    plt.show()
    print('\n')
    plt.clf()
```
